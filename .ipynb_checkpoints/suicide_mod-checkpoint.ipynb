This notebook builds up the core components of the general escape theory of suicide step-by-step. We start with simulations of (1) stressors only, and then incrementally add in (2) negative affect, (3) urge to escape, (4) suicidal thoughts, (5) short-term escape behaviors, and finally present the full model including (6) long-term change. 

At each step of this process, we observe the simulations to determine whether they can adequately reproduce features of the phenomena that we observe in the real world. For instance, the simulations should produce suicidal thoughts that are zero-inflated, have a fast onset, and have a short duration, because this is what we have consistently observed in ecological momentary assessment studies of suicidal thoughts. 

We also observe the simulations to determine whether they can reproduce features of the phenomena that are in line with our theory. For instance, we should be able to manipulate parameters such that suicidal thoughts do *not* emerge when short-term escape behaviors are effectively regulating negative affect, and only emerge when short-term escape behaviors no longer effectively reduce negative affect.

## 1. Stressors only

Geometric Brownian Motion - equation from https://en.wikipedia.org/wiki/Geometric_Brownian_motion. 

$S_t = S_0 e^{(\mu - \frac{\sigma^2}{2})t + \sigma W_t}$

Desired features of phenomenon: 

- In general, stressors should remain relatively low, with small fluctuations up and down.
- However, stressors *should* sometimes show sharp spikes up and down.
- There should be a fair amount of variability in stressors over time.
- Stressors should be above zero.

```python
def geometric_brownian(t, dt, mu, sigma, s0):
    '''
    args:
        t: time period to integrate over
        dt: time step
        mu: drift
        sigma: volatility
        s0: initial value
    '''
    
    stressor = np.zeros(t)
    stressor[0] = s0
    
    for i in range(0, len(stressor) - 1):
        stressor[i + 1] = stressor[i]*np.exp((mu - ((sigma**2)/2))*dt + sigma*np.random.normal(0, np.sqrt(dt)))
    
    return(stressor)
```

```python
## figure parameters 
fig_width = 10
fig_height = 7
legend_size = 10
axis_size = 14
```

```python
# plot geometric brownian motion
t = 10000
dt = 0.01
mu = 0.00
sigma = 0.15
s0 = 0.2

fig, ax = plt.subplots(1, 1, figsize = (10, 5))
for i in range(10):
    stressor = geometric_brownian(t, dt, mu, sigma, s0)
    ax.plot(stressor)
    ax.set_ylim([0, 1])
fig.text(0.5, 0.02, 'Time', ha='center', fontsize = axis_size)
fig.text(0.06, 0.5, 'Intensity', va='center', rotation='vertical', fontsize = axis_size)
#plt.show()    
#plt.title(r'10 Realizations of Stressor: $S_t = S_0 e^{(\mu - \frac{\sigma^2}{2})t + \sigma W_t}$');
plt.savefig("figures/foryale/stressor.png", dpi = 300, transparent = True);
```

## 2. Stressor & Negative Affect

Stressors: $S_t = S_0 e^{(\mu - \frac{\sigma^2}{2})t + \sigma W_t}$

Negative Affect: $\frac{dN}{dt} = b_2 N(K-N) + a_2S$

Desired features of phenomena: 

- In general, negative affect should closely follow the stressors. When stressors increase, negative affect should increase, and when stressors decrease, negative affect should decrease. 
- In the absense of stressors, negative affect should trend towards a person's "carrying capacity".

```python
def gen_stress_na(t, dt, mu, sigma, s0, n0, b2, K, a2):
    '''
    args:
        t: time period to integrate over
        s0: initial value stressor 
        dt: time step 
        a1: 'speed' of Brownian motion 
        n0: initial value NA
        b2: slope of NA growth rate
        K: carrying capacity 
        a2: effect of stressor --> dN/dt
    
    returns:
        np.array of shape (2, t)
    '''
    
    stress_na = np.zeros([2, t])
    stress_na[0] = geometric_brownian(t, dt, mu, sigma, s0)
    stress_na[1][0] = n0
    
    for i in range(t - 1):
        stress_na[1][i + 1] = stress_na[1][i] + dt*(b2*stress_na[1][i]*(K - stress_na[1][i]) + a2*stress_na[0][i])
    
    return(stress_na)
```

```python
# time and time step
t, dt = 10000, 0.01

# stressor params - geometric brownian motion
mu, sigma, s0 = 0.0, 0.15, 0.1

# neg affect params
# b2 = growth rate, a2 = effect of stressor --> NA; K = carrying capacity 
n0, b2, K, a2 = 0.3, 2, 0.2, 1.5

fig, ax = plt.subplots(2, 2, figsize=(fig_width,fig_height), sharey = False)
ax = ax.ravel()
for i in range(4):
    stress_na = gen_stress_na(t, dt, mu, sigma, s0, n0, b2, K, a2)
    ax[i].plot(stress_na[0], label = 'stressor')
    ax[i].plot(stress_na[1], label = 'negative affect')
    ax[i].set_ylim([0, 1])
    if i == 0:
        fig.legend(bbox_to_anchor=(0.325, 0.885), fontsize = legend_size)
fig.text(0.5, 0.05, 'Time', ha='center', fontsize = axis_size)
fig.text(0.06, 0.5, 'Intensity', va='center', rotation='vertical', fontsize = axis_size)
#plt.show()
#plt.suptitle('4 Realizations of Stressor and Negative Affect', fontsize = 16);
plt.savefig("figures/foryale/stressor_na.png", dpi = 300, transparent = True, bbox_inches='tight');
```

## 3. Stressor & Negative Affect & Urge to Escape

Stressors: $S_t = S_0 e^{(\mu - \frac{\sigma^2}{2})t + \sigma W_t}$

Negative Affect: $\frac{dN}{dt} = b_2 N(K-N) + a_2S$ 

Urge to Escape: $\frac{dU}{dt} = -c_3U + b_3N$

Desired features of phenomena: 

- In general, urge to escape should follow negative affect. When negative affect increases, urge to escape should increase, and when negative affect decreases, urge to escape should decrease.


```python
def gen_stress_na_urge(t, dt, mu, sigma, s0, n0, b2, K, a2, u0, b3, c3):
    '''
    args:
        t: time period to integrate over
        s0: initial value stressor 
        dt: time step 
        a1: 'speed' of Brownian motion 
        n0: initial value NA
        b2: slope of NA growth rate
        K: carrying capacity 
        a2: effect of stressor --> dN/dt
        u0: initial value urge to escape
        b3: effect of NA --> dU/dt
        c3: natural decay of urge to escape
    
    returns:
        np.array of shape (3, t)
    '''
    
    stress_na_urge = np.zeros([3, t])
    stress_na_urge[0:2, ] = gen_stress_na(t, dt, mu, sigma, s0, n0, b2, K, a2)
    stress_na_urge[2][0] = u0
    
    for i in range(t - 1):
        stress_na_urge[2][i + 1] = dt*(-c3*stress_na_urge[2][i] + b3*stress_na_urge[1][i]) + stress_na_urge[2][i]
    
    return(stress_na_urge)
```

```python
# time and time step
t, dt = 10000, 0.01

# stressor params - geometric brownian motion
mu, sigma, s0 = 0.0, 0.15, 0.1

# neg affect params
# b2 = growth rate, a2 = effect of stressor --> NA; K = carrying capacity 
n0, b2, K, a2 = 0.3, 2, 0.2, 1.5

# urge to escape params
# b3 = influence of NA --> urge; c3 = urge self-decay
u0, b3, c3 = 0.1, 2, 3

fig, ax = plt.subplots(2, 2, figsize=(fig_width,fig_height), sharey = False)
ax = ax.ravel()
for i in range(4): 
    stress_na_urge = gen_stress_na_urge(t, dt, mu, sigma, s0, n0, b2, K, a2, u0, b3, c3)
    ax[i].plot(stress_na_urge[0], label = 'stressor')
    ax[i].plot(stress_na_urge[1], label = 'negative affect')
    ax[i].plot(stress_na_urge[2], label = 'urge to escape')
    #ax[i].legend(loc = 'upper left')
    ax[i].set_ylim([0, 1])
    if i == 0:
        fig.legend(bbox_to_anchor=(0.325, 0.885), fontsize = legend_size)
fig.text(0.5, 0.05, 'Time', ha='center', fontsize = axis_size)
fig.text(0.06, 0.5, 'Intensity', va='center', rotation='vertical', fontsize = axis_size)
#plt.show()
#plt.suptitle('4 Realizations of Stressor and Negative Affect', fontsize = 16);
plt.savefig("figures/foryale/stressor_na_urge.png", dpi = 300, transparent = True, bbox_inches='tight');
```

## 4. Stressor & Negative Affect & Urge to Escape & Suicidal Thoughts

Stressors: $S_t = S_0 e^{(\mu - \frac{\sigma^2}{2})t + \sigma W_t}$

Negative Affect: $\frac{dN}{dt} = b_2 N(K-N) + a_2S - d_2T$

Urge to Escape: $\frac{dU}{dt} = -c_3U + b_3N$

Suicidal Thoughts: $\frac{dT}{dt} = -d_4 T + \frac{1}{1 + e^{-c_{41}(U - c_{42})}}$

Desired features of phenomena: 

- Unlike negative affect and urge to escape, suicidal thoughts shold *not* directly follow stressors. 
- Given what we know from EMA studies, suicidal thoughts should have a relatively fast onset, and short duration. 
- In addition, given what we know from EMA studies, suicidal thoughts should be zero-inflated.

```python
def gen_stress_na_urge_si(t, dt, mu, sigma, s0, n0, b2, K, a2, u0, b3, c3, t0, c41, c42, d4, d2):
    '''
    args:
        t: time period to integrate over
        s0: initial value stressor 
        dt: time step 
        a1: 'speed' of Brownian motion 
        n0: initial value NA
        b2: slope of NA growth rate
        K: carrying capacity 
        a2: effect of stressor --> dN/dt
        u0: initial value urge to escape
        b3: effect of NA --> dU/dt
        c3: natural decay of urge to escape
        t0: initial value of suicidal thoughts
        c41: effect of urge to escape --> dS/dt; steepness of logistic curve
        c42: midpoint of logistic curve
        d4: natural decay of SI over time
        d2: effect of SI --> -dN/dt
    
    returns:
        np.array of shape (4, t)
    '''
    
    stress_na_urge_si = np.zeros([4, t])
    stress_na_urge_si[0] = geometric_brownian(t, dt, mu, sigma, s0)
    stressor = stress_na_urge_si[0]
    
    neg_affect = stress_na_urge_si[1]
    urge_escape = stress_na_urge_si[2]
    sui_thoughts = stress_na_urge_si[3]
    
    neg_affect[0] = n0
    urge_escape[0] = u0
    sui_thoughts[0] = t0
    
    for i in range(t - 1):
        ## negative affect including negative effect of suicidal thoughts
        neg_affect[i + 1] = neg_affect[i] + dt*(b2*neg_affect[i]*(K - neg_affect[i]) + a2*stressor[i] - d2*sui_thoughts[i])
        
        ## urge to escape
        urge_escape[i + 1] = urge_escape[i] + dt*(-c3*urge_escape[i] + b3*neg_affect[i]) 
        
        ## suicidal thoughts
        sui_thoughts[i + 1] = sui_thoughts[i] + dt*(-d4*sui_thoughts[i] + (1 / (1 + np.exp(-c41 * (urge_escape[i] - c42)))))
    
    return(stress_na_urge_si)
```

```python
# time and time step
t, dt = 10000, 0.01

# stressor params - geometric brownian motion
mu, sigma, s0 = 0.0, 0.15, 0.1

# neg affect params
# b2 = growth rate, a2 = effect of stressor --> NA; K = carrying capacity 
n0, b2, K, a2 = 0.3, 2, 0.2, 1.5

# urge to escape params
# b3 = influence of NA --> urge; c3 = urge self-decay
u0, b3, c3 = 0.1, 2, 3

# suicidal thoughts params
# c41 = steepness; c42 = midpoint; d4 = self-decay; d2 = effect on NA 
t0, c41, c42, d4, d2 = 0, 55, 0.35, 1, 0.1

fig, ax = plt.subplots(2, 2, figsize=(fig_width,fig_height), sharey = False)
ax = ax.ravel()
for i in range(4): 
    stress_na_urge_si = gen_stress_na_urge_si(t, dt, mu, sigma, s0, n0, b2, K, a2, u0, b3, c3, t0, c41, c42, d4, d2)
    ax[i].plot(stress_na_urge_si[0], label = 'stressor')
    ax[i].plot(stress_na_urge_si[1], label = 'negative affect')
    ax[i].plot(stress_na_urge_si[2], label = 'urge to escape')
    ax[i].plot(stress_na_urge_si[3], label = 'suicidal thoughts')
    #ax[i].legend(loc = 'upper left')
    ax[i].set_ylim([0, 1])
    if i == 0:
        fig.legend(bbox_to_anchor=(0.308, 0.882), fontsize = 10)
fig.text(0.5, 0.05, 'Time', ha='center', fontsize = axis_size)
fig.text(0.06, 0.5, 'Intensity', va='center', rotation='vertical', fontsize = axis_size)
#plt.show()
#plt.suptitle('4 Realizations of Stressor and Negative Affect', fontsize = 16);
plt.savefig("figures/foryale/stressor_na_urge_si.png", dpi = 300, transparent = True, bbox_inches='tight');
```

## 5. Adding other short-term escape behaviors

Stressors: $S_t = S_0 e^{(\mu - \frac{\sigma^2}{2})t + \sigma W_t}$

Negative Affect: $\frac{dN}{dt} = b_2 N(K-N) + a_2S - d_2T - e_2O$

Urge to Escape: $\frac{dU}{dt} = -c_3U + b_3N$

Suicidal Thoughts: $\frac{dT}{dt} = -d_4 T + \frac{1}{1 + e^{-c_{41}(U - c_{42})}}$

Other Short-Term Escape: $\frac{dO}{dt} = -e_5O + \frac{1}{1 + e^{-c_{51}(U - c_{52})}}$

Desired feature of phenomena: 

- A core component of our theory (General Escape Theory of Suicide) is that suicidal thoughts should *not* emerge when other escape behaviors are effectively regulating negative affect. 
- Therefore, when the parameter governing the effect of other escape behaviors on change in negative affect over time ($e_2$ parameter) is *large*, suicidal thoughts should not emerge.
- On the other hand, when the $e_2$ parameter is *small*, suicidal thoughts should emerge.


```python
def gen_stress_na_urge_si_other(t, dt, mu, sigma, s0, n0, b2, K, a2, u0, b3, c3, t0, c41, c42, d4, d2, o0, e5, c51, c52, e2):
    '''
    args:
        t: time period to integrate over
        s0: initial value stressor 
        dt: time step 
        a1: 'speed' of Brownian motion 
        n0: initial value NA
        b2: slope of NA growth rate
        K: carrying capacity 
        a2: effect of stressor --> dN/dt
        u0: initial value urge to escape
        b3: effect of NA --> dU/dt
        c3: natural decay of urge to escape
        t0: initial value of suicidal thoughts
        c41: effect of urge to escape --> dS/dt; steepness of logistic curve
        c42: midpoint of logistic curve
        d4: natural decay of SI over time
        d2: effect of SI --> -dN/dt
        e5: natural decay of other short-term escape over time
        c51: effect of urge to escape --> dO/dt; steepness of logistic curve
        c52: midpoint of logistic curve
        e2: effect of other short-term escape --> -dN/dt
    
    returns:
        np.array of shape (5, t)
    '''
    
    stress_na_urge_si_oth = np.zeros([5, t])
    stress_na_urge_si_oth[0] = geometric_brownian(t, dt, mu, sigma, s0)
    stressor = stress_na_urge_si_oth[0]
    
    neg_affect = stress_na_urge_si_oth[1]
    urge_escape = stress_na_urge_si_oth[2]
    sui_thoughts = stress_na_urge_si_oth[3]
    other_escape = stress_na_urge_si_oth[4]
    
    neg_affect[0] = n0
    urge_escape[0] = u0
    sui_thoughts[0] = t0
    other_escape[0] = o0
    
    for i in range(t - 1):
        ## negative affect including negative effect of suicidal thoughts
        neg_affect[i + 1] = neg_affect[i] + dt*(b2*neg_affect[i]*(K - neg_affect[i]) + a2*stressor[i] - d2*sui_thoughts[i] - e2*other_escape[i])
        
        ## urge to escape
        urge_escape[i + 1] = urge_escape[i] + dt*(-c3*urge_escape[i] + b3*neg_affect[i])
        
        ## suicidal thoughts
        sui_thoughts[i + 1] = sui_thoughts[i] + dt*(-d4*sui_thoughts[i] + (1 / (1 + np.exp(-c41 * (urge_escape[i] - c42)))))
        
        ## other escape behaviors 
        other_escape[i + 1] = other_escape[i] + dt*(-e5*other_escape[i] + (1 / (1 + np.exp(-c51 * (urge_escape[i] - c52)))))
    
    return(stress_na_urge_si_oth)
```

```python
# time and time step
t, dt = 10000, 0.01

# stressor params - geometric brownian motion
mu, sigma, s0 = 0.0, 0.15, 0.1

# neg affect params
# b2 = growth rate, a2 = effect of stressor --> NA; K = carrying capacity 
n0, b2, K, a2 = 0.3, 5, 0.2, 3

# urge to escape params
# b3 = influence of NA --> urge; c3 = urge self-decay
u0, b3, c3 = 0.1, 2, 3

# suicidal thoughts params
# c41 = steepness; c42 = midpoint; d4 = self-decay; d2 = effect on NA 
t0, c41, c42, d4, d2 = 0, 55, 0.45, 1, 0.1

# other escape params
# e5 = self-decay; c51 = steepness; c52 = midpoint; e2 = effect on NA 
o0, e5, c51, c52, e2 = 0, 2, 35, 0.3, 4.3

fig, ax = plt.subplots(2, 2, figsize=(fig_width,fig_height), sharey = False)
ax = ax.ravel()
for i in range(4): 
    stress_na_urge_si_other = gen_stress_na_urge_si_other(t = t, dt = dt, mu = mu, sigma = sigma, s0 = s0, 
                                                          n0 = n0, b2 = b2, K = K, a2 = a2, u0 = u0, b3 = b3, c3 = c3, 
                                                          t0 = t0, c41 = c41, c42 = c42, d4 = d4, d2 = d2, 
                                                          o0 = o0, e5 = e5, c51 = c51, c52 = c52, e2 = e2)
    ax[i].plot(stress_na_urge_si_other[0], label = 'stressor')
    ax[i].plot(stress_na_urge_si_other[1], label = 'negative affect')
    ax[i].plot(stress_na_urge_si_other[2], label = 'urge to escape')
    ax[i].plot(stress_na_urge_si_other[3], label = 'suicidal thoughts')
    ax[i].plot(stress_na_urge_si_other[4], label = 'other escape')
    #ax[i].legend(loc = 'upper left')
    ax[i].set_ylim([0, 1])
    if i == 0:
        fig.legend(bbox_to_anchor=(0.31, 0.88), fontsize = 10)
fig.text(0.5, 0.05, 'Time', ha='center', fontsize = axis_size)
fig.text(0.06, 0.5, 'Intensity', va='center', rotation='vertical', fontsize = axis_size)
#plt.show()
#plt.suptitle('4 Realizations of Stressor and Negative Affect', fontsize = 16);
plt.savefig("figures/foryale/stressor_na_urge_si_oth_eff.png", dpi = 300, transparent = True, bbox_inches='tight');
```

```python
# other escape params
# e2 = effect on NA 
e2 = 0.00000001

fig, ax = plt.subplots(2, 2, figsize=(fig_width,fig_height), sharey = False)
ax = ax.ravel()
for i in range(4): 
    stress_na_urge_si_other = gen_stress_na_urge_si_other(t = t, dt = dt, mu = mu, sigma = sigma, s0 = s0, 
                                                          n0 = n0, b2 = b2, K = K, a2 = a2, u0 = u0, b3 = b3, c3 = c3, 
                                                          t0 = t0, c41 = c41, c42 = c42, d4 = d4, d2 = d2, 
                                                          o0 = o0, e5 = e5, c51 = c51, c52 = c52, e2 = e2)
    ax[i].plot(stress_na_urge_si_other[0], label = 'stressor')
    ax[i].plot(stress_na_urge_si_other[1], label = 'negative affect')
    ax[i].plot(stress_na_urge_si_other[2], label = 'urge to escape')
    ax[i].plot(stress_na_urge_si_other[3], label = 'suicidal thoughts')
    ax[i].plot(stress_na_urge_si_other[4], label = 'other escape')
    #ax[i].legend(loc = 'upper left')
    ax[i].set_ylim([0, 1])
    if i == 0:
        fig.legend(bbox_to_anchor=(0.31, 0.88), fontsize = 10)
fig.text(0.5, 0.05, 'Time', ha='center', fontsize = axis_size)
fig.text(0.06, 0.5, 'Intensity', va='center', rotation='vertical', fontsize = axis_size)
#plt.show()
#plt.suptitle('4 Realizations of Stressor and Negative Affect', fontsize = 16);
plt.savefig("figures/foryale/other_noteff.png", dpi = 300, transparent = True, bbox_inches='tight');
```

## 6. Adding long-term change
    
Stressors: $S_t = S_0 e^{(\mu - \frac{\sigma^2}{2})t + \sigma W_t} e^{ - f_1L}$

Negative Affect: $\frac{dN}{dt} = b_2 N(K_2-N) + a_2S - d_2T - e_2O$

Urge to Escape: $\frac{dU}{dt} = -c_3U + b_3N$

Suicidal Thoughts: $\frac{dT}{dt} = -d_4 T + \frac{1}{1 + e^{-c_{41}(U - c_{42})}}$

Other Escape Behavior: $\frac{dO}{dt} = -e_5O + \frac{1}{1 + e^{-c_{51}(U - c_{52})}}$

Long-Term Change: $\frac{dL}{dt} = f_6L(K_6-L) + b_6N - c_6U$
    

#### Simulations and expected behavior (theorized phenomena we hope to reproduce): 

- Other escape behaviors effectively reduce NA; suicidal thoughts don't emerge [*replication of above*]:  $e_2$ parameter
- Other escape behaviors do **not** effectively reduce NA; suicidal thoughts emerge [*replication of above*]: $e_2$ parameter
- Long-term change effectively reduces stressor; low NA; no other escape/SI emerge: $f_1$ parameter
- Long-term change does **not** effectively reduce stressor; high NA; other escape/SI emerge: $f_1$ parameter
- Low engagement in long-term change; high NA; other escape/SI emerge: $K_6, b_6, c_6$ parameters
- Negative affect has a weak effect on urge to escape; suicidal thoughts and other escape behaviors do not emerge much: $b_3$ parameter

```python
def gen_stress_na_urge_si_other_lt(t, dt, mu, sigma, s0, n0, b2, K2, a2, u0, b3, c3, t0, c41, c42, d4, d2, o0, e5, c51, c52, e2, l0, f6, b6, f1, c6, K6):
    
    '''
    new args:
        l0: initial value long-term change
        f6: self-increase of long-term change
        c6: effect of urge to escape --> long-term change
        K6: crarying capacity for long-term change
    
    returns:
        np.array of shape (6, t)
    '''
    
    stress_na_urge_si_oth_lt = np.zeros([6, t])
    
    stressor = stress_na_urge_si_oth_lt[0]
    neg_affect = stress_na_urge_si_oth_lt[1]
    urge_escape = stress_na_urge_si_oth_lt[2]
    sui_thoughts = stress_na_urge_si_oth_lt[3]
    other_escape = stress_na_urge_si_oth_lt[4]
    long_term = stress_na_urge_si_oth_lt[5]
    
    stressor[0] = s0
    neg_affect[0] = n0
    urge_escape[0] = u0
    sui_thoughts[0] = t0
    other_escape[0] = o0
    long_term[0] = l0
            
    for i in range(t - 1):
            
        ## stressor and effect of long-term change
        stressor[i + 1] = stressor[i]*np.exp((mu - ((sigma**2)/2))*dt + sigma*np.random.normal(0, np.sqrt(dt)) - f1*long_term[i-1])
        
        ## negative affect including negative effect of suicidal thoughts
        neg_affect[i + 1] = neg_affect[i] + dt*(b2*neg_affect[i]*(K2 - neg_affect[i]) + a2*stressor[i] - d2*sui_thoughts[i] - e2*other_escape[i])
        
        ## urge to escape
        urge_escape[i + 1] = urge_escape[i] + dt*(-c3*urge_escape[i] + b3*neg_affect[i]) # + 5*dt*new_effect OR really weak new_effect
        
        ## suicidal thoughts
        sui_thoughts[i + 1] = sui_thoughts[i] + dt*(-d4*sui_thoughts[i] + (1 / (1 + np.exp(-c41 * (urge_escape[i] - c42)))))
        
        ## other escape behaviors 
        other_escape[i + 1] = other_escape[i] + dt*(-e5*other_escape[i] + (1 / (1 + np.exp(-c51 * (urge_escape[i] - c52)))))
        
        ## long-term change
        long_term[i + 1] = long_term[i] + dt*(f6*long_term[i]*(K6 - long_term[i]) + b6*neg_affect[i] - c6*urge_escape[i])
    
    return(stress_na_urge_si_oth_lt)
```

```python
# time and time step
t, dt = 10000, 0.01

# stressor params - geometric brownian motion
mu, sigma, s0 = 0.0, 0.15, 0.1

# neg affect params
# b2 = growth rate, a2 = effect of stressor --> NA; K = carrying capacity 
n0, b2, K2, a2 = 0.3, 5, 0.2, 3

# urge to escape params
# b3 = influence of NA --> urge; c3 = urge self-decay
u0, b3, c3 = 0.1, 2, 3

# suicidal thoughts params
# c41 = steepness; c42 = midpoint; d4 = self-decay; d2 = effect on NA 
t0, c41, c42, d4, d2 = 0, 50, 0.4, 0.5, 0.01

# other escape params
# e5 = self-decay; c51 = steepness; c52 = midpoint; e2 = effect on NA 
o0, e5, c51, c52, e2 = 0, 1.5, 45, 0.3, 6

# long-term change params
# f6 = self-growth; b6 = influence of NA --> long-term change; f1 = effect on stressor; c6 = influence of urge --> LT change
l0, K6, f6, b6, f1, c6 = 0.1, 0.2, 0.2, 2, 0.000001, 3

fig, ax = plt.subplots(2, 2, figsize=(fig_width,fig_height), sharey = False)
ax = ax.ravel()
for i in range(4): 
    stress_na_urge_si_other_lt = gen_stress_na_urge_si_other_lt(t = t, dt = dt, mu = mu, sigma = sigma, s0 = s0, 
                                                                n0 = n0, b2 = b2, K2 = K2, a2 = a2, u0 = u0, b3 = b3, c3 = c3, 
                                                                t0 = t0, c41 = c41, c42 = c42, d4 = d4, d2 = d2, 
                                                                o0 = o0, e5 = e5, c51 = c51, c52 = c52, e2 = e2,
                                                                l0 = l0, f6 = f6, b6 = b6, f1 = f1, c6 = c6, K6 = K6)
    ax[i].plot(stress_na_urge_si_other_lt[0], label = 'stressor')
    ax[i].plot(stress_na_urge_si_other_lt[1], label = 'negative affect')
    ax[i].plot(stress_na_urge_si_other_lt[2], label = 'urge to escape')
    ax[i].plot(stress_na_urge_si_other_lt[3], label = 'suicidal thoughts')
    ax[i].plot(stress_na_urge_si_other_lt[4], label = 'other short-term escape')
    ax[i].plot(stress_na_urge_si_other_lt[5], label = 'long-term change')
    ax[i].set_ylim([0, 1])
    if i == 0:
        fig.legend(bbox_to_anchor=(1.14, 0.9), fontsize = legend_size)
fig.text(0.5, 0.02, 'Time', ha='center', fontsize = axis_size)
fig.text(0.06, 0.5, 'Intensity', va='center', rotation='vertical', fontsize = axis_size)

#fig.suptitle("6a. Short-Term Escape Bx Effective; No Suicidal Thoughts Emerge", size = 16)
plt.show()
```
